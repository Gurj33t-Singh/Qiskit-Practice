wave plot with λ=1m,
ω=1s which means 1 sine wave or 2π raidans in 1 second
k=2π radians/meter
A=1 (amplitude)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0            # Amplitude
k = 2 * np.pi       # Wave number (1 m wavelength => k = 2π)
omega = 2 * np.pi   # Angular frequency (1 s period)

# Space and Time
x = np.linspace(0, 2, 1000)  # 2 meters
t = 0.25                      # fixed time point for snapshot

# Wave
psi = A * np.exp(-1j * (k * x - omega * t))
psi_real = np.real(psi)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(x, psi_real, label=r'Re($e^{-i(kx-\omega t)}$)')
plt.title(f'Wave Snapshot at t = {t}s')
plt.xlabel('Position (m)')
plt.ylabel('Wave Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0           # Amplitude
k = 2 * np.pi      # Wave number (1m wavelength => k = 2π)
omega = 2 * np.pi  # Angular frequency (1s period)

# Space range
x = np.linspace(0, 2, 1000)

# Times to snapshot
times = [0.0, 0.25, 0.5, 0.75]

# Plot
plt.figure(figsize=(10, 6))
for t in times:
    psi = A * np.cos(k * x - omega * t)  # Real part
    plt.plot(x, psi, label=f't = {t}s')

plt.title('Wave Propagation at Multiple Times')
plt.xlabel('Position (m)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definitions
t = np.linspace(0, 2*np.pi, 1000)  # Angle from 0 to 2π
A = 1.0

# Waveforms
sin_wave = A * np.sin(t)
cos_wave = A * np.cos(t)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, sin_wave, label='Sine wave (starts at 0)')
plt.plot(t, cos_wave, label='Cosine wave (starts at A)')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(np.pi/2, color='gray', linestyle='--')
plt.axvline(np.pi, color='gray', linestyle='--')
plt.axvline(3*np.pi/2, color='gray', linestyle='--')
plt.axvline(2*np.pi, color='gray', linestyle='--')
plt.legend()
plt.title('Sine vs Cosine Waves')
plt.xlabel('Angle (radians)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Parameters
t_values = np.linspace(0, 2*np.pi, 100)

# Set up figure
fig, ax = plt.subplots(figsize=(5,5))
ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)

# Elements
point, = ax.plot([], [], 'o', color='red')
line_in_phase, = ax.plot([], [], color='blue', label='In-Phase (cos)')
line_quadrature, = ax.plot([], [], color='green', label='Quadrature (sin)')

# Labels
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.legend()

def animate(i):
    t = t_values[i]
    x = np.cos(t)  # In-Phase
    y = np.sin(t)  # Quadrature
    point.set_data([x], [y])
    line_in_phase.set_data([0, x], [0, 0])     # cos -> X direction
    line_quadrature.set_data([0, 0], [0, y])    # sin -> Y direction
    ax.set_title(f'Time = {t:.2f} | cos(t) = {x:.2f}, sin(t) = {y:.2f}')
    return point, line_in_phase, line_quadrature

ani = FuncAnimation(fig, animate, frames=len(t_values), interval=100, blit=True)

plt.show()
```
